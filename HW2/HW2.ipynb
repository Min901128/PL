```python
import json
```

```python
import pprint as pp

with open('111年-舉發交通違規明細表.json', encoding='utf-8') as file:
    data = json.load(file)

pp.pprint(data, sort_dicts=False)
```

```python
# Q1, 查詢某法條的最高裁罰金額
found = False

x = input("您想查詢第幾號法條：")
Target = "第" + x + "條"

for item in data:
    if Target in item['違規罰單法條']:
        print(Target + f"的最高裁罰金額為{item['最高裁罰金額']}")
        found = True

if not found:
    print("找不到該法條的資料")
```

```python
# Q2, 查詢哪些法條的最高裁罰金額超過??元?
found = False
first = False

money = int(input("請輸入金額："))#這邊要轉數字

for item in data:
    if item['最高裁罰金額'] >= money:
        if first == False:
            print("超過" + str(money) + "的有以下:")
            first = True
        
        print(f"{item['違規罰單法條']}的最高裁罰金額為{item['最高裁罰金額']}")
        found = True

if not found:
    print("找不到超過該金額的資料")
```

```python
# Q3, 哪個法條的裁罰Range最大
FirstItem = data[0]
LargestRange = FirstItem['最高裁罰金額'] - FirstItem['最低裁罰金額']
LargestRangeLaw = FirstItem['違規罰單法條']

for item in data:
    Range = item['最高裁罰金額'] - item['最低裁罰金額']
    print(f"{item['違規罰單法條']}的Range為" + str(Range)) #print各法條的Range
    if Range >= LargestRange:
        LargestRange = Range
        LargestRangeLaw = item['違規罰單法條']
print()
print("Range最大的法條為" + LargestRangeLaw)
```
