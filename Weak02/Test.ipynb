```python
# Data1
import pandas as pd
# 顯示資料
data1=pd.read_csv("各廠牌大型重機新車領牌數.csv", encoding='big5')
print(data1)
```

```python
# Data1篩選出112年且廠牌為日系四大廠的數據
data1 = data1[data1['廠牌'].isin(['HONDA', 'KAWASAKI', 'SUZUKI', 'YAMAHA'])]
data1 = data1[data1['年份'].isin(['112年'])]
print(data1)
```

```python
# Q1, 112年日系四大廠大型重機新車領牌數的情況
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.ticker as ticker  
import matplotlib.pyplot as plt

# 按品牌拆分數據
honda_data = data1[data1['廠牌'] == 'HONDA']
yamaha_data = data1[data1['廠牌'] == 'YAMAHA']
suzuki_data = data1[data1['廠牌'] == 'SUZUKI']
kawasaki_data = data1[data1['廠牌'] == 'KAWASAKI']

x = [1, 2, 3, 4, 5,6,7,8]
y_honda = honda_data['數量']
y_yamaha = yamaha_data['數量']
y_suzuki = suzuki_data['數量']
y_kawasaki = kawasaki_data['數量']

fig, ax1 = plt.subplots() #創建圖形

# 第1條線
ax1.plot(x, y_honda, label='HONDA', linestyle='-', marker='o', color='Red') # marker='o' 圓形標誌
ax1.set_ylabel('Mount')

# 第2條線
ax2 = ax1.twinx()
ax2.plot(x, y_yamaha, label='YAMAHA', linestyle='-', marker='s', color='Blue')

# 第3條線
ax3 = ax1.twinx()
ax3.plot(x, y_suzuki, label='SUZUKI', linestyle='-', marker='^', color='PURPLE')

# 第4條線
ax4 = ax1.twinx()
ax4.plot(x, y_kawasaki, label='KAWASAKI', linestyle='-', marker='x', color='Green')

# 添加圖例在右上角
fig.legend(loc="upper right")

# 設置横軸標籤
ax1.set_xlabel('Month')

plt.show()
```
