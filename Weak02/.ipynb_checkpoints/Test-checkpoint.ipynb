```python
import pandas as pd
# 顯示資料
data=pd.read_csv("大重領牌情況.csv", encoding='big5')
print(data)
```

```python
import matplotlib.pyplot as plt
x=list(data["統計期"])
y=list(data["大重總計"])
print(x)
print(y)
```

```python
import matplotlib.ticker as ticker  # ticker模塊

plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))  # 格式化為整數，逗號分隔
plt.plot(x,y)
plt.show()
```
