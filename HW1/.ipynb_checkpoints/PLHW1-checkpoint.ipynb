```python
# Data1
import pandas as pd
# 顯示資料
data1=pd.read_csv("各廠牌大型重機新車領牌數.csv", encoding='big5')
print(data1)
```

```python
# Data1篩選出112年且廠牌為日系四大廠的數據
data1 = data1[data1['廠牌'].isin(['HONDA', 'KAWASAKI', 'SUZUKI', 'YAMAHA'])]
data1 = data1[data1['年份'].isin(['112年'])]
print(data1)
```

```python
# Data2
import pandas as pd
# 顯示資料
data2=pd.read_csv("各廠牌大型重機車輛登記數.csv", encoding='big5')
print(data2)
```

```python
# Data3
import pandas as pd
# 顯示資料
data3=pd.read_csv("大重領牌情況.csv", encoding='big5')
print(data3)
```

```python
# Data4
import pandas as pd
# 顯示資料
data4=pd.read_csv("普重領牌情況.csv", encoding='big5')
print(data4)
```

```python
# Data5
import pandas as pd
# 顯示資料
data5=pd.read_csv("小客車領牌情況.csv", encoding='big5')
print(data5)
```

```python
# Q1, 112年日系四大廠大型重機新車領牌數的情況
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.ticker as ticker  
import matplotlib.pyplot as plt

# 按品牌拆分數據
honda_data = data1[data1['廠牌'] == 'HONDA']
yamaha_data = data1[data1['廠牌'] == 'YAMAHA']
suzuki_data = data1[data1['廠牌'] == 'SUZUKI']
kawasaki_data = data1[data1['廠牌'] == 'KAWASAKI']

x = [1, 2, 3, 4, 5,6,7,8]
y_honda = honda_data['數量']
y_yamaha = yamaha_data['數量']
y_suzuki = suzuki_data['數量']
y_kawasaki = kawasaki_data['數量']

fig, ax1 = plt.subplots() #創建圖形

plt.title('112年日系四大廠大型重機新車領牌數的情況')

# 第1條線
ax1.plot(x, y_honda, label='HONDA', linestyle='-', marker='o', color='Red') # marker='o' 圓形標誌
ax1.set_ylabel('數量')

# 第2條線
ax2 = ax1.twinx()
ax2.plot(x, y_yamaha, label='YAMAHA', linestyle='-', marker='s', color='Blue')

# 第3條線
ax3 = ax1.twinx()
ax3.plot(x, y_suzuki, label='SUZUKI', linestyle='-', marker='^', color='PURPLE')

# 第4條線
ax4 = ax1.twinx()
ax4.plot(x, y_kawasaki, label='KAWASAKI', linestyle='-', marker='x', color='Green')

# 添加圖例在右上角
fig.legend(loc="upper right")

# 設置横軸標籤
ax1.set_xlabel('月份')

plt.show()
```

```python
# Q2, 112年8月Honda新車領牌數與該廠牌車輛登記數之比較
# 篩選Honda 112年8月的數據
filtered_data = data1[(data1['統計期'].str.contains('112年 8月')) & (data1['廠牌'].str.contains('HONDA'))]
filtered_data = filtered_data.rename(columns={'數量': '領牌數'})
filtered_data2 = data2[(data2['統計期'].str.contains('112年 8月')) & (data2['廠牌'].str.contains('HONDA'))]
filtered_data2 = filtered_data2.rename(columns={'數量': '登記數'})

# 合併数据
merged_df = pd.merge(filtered_data, filtered_data2, on='廠牌')

# 使用melt函数將數據重排
merged_df = pd.melt(merged_df, id_vars=['廠牌'], var_name='類別', value_name='數量')

# 篩選類別為登記數或領牌數的資料
merged_df = merged_df[(merged_df['類別'].str.contains('登記數')) | (merged_df['類別'].str.contains('領牌數'))]

x = merged_df['類別']
y = merged_df['數量']

# 創建一個長條圖
plt.figure(figsize=(10, 6))  # 設置圖形大小
plt.bar(x, y, color='Blue')
plt.title('112年8月Honda新車領牌數與該廠牌車輛登記數之比較')
plt.xlabel('類別')
plt.ylabel('數量')

# 顯示圖表
plt.tight_layout()
plt.show()
```

```python
# Q3, 112年8月大重普重小客車的領牌情況

# 合併數據
merged_df = pd.merge(data3, data4, on='統計期')
merged_df2 = pd.merge(merged_df, data5, on='統計期')

# 使用melt函数將數據重排
merged_df2 = pd.melt(merged_df2, id_vars=['統計期'], var_name='類別', value_name='數量')

# 篩選112年8月的數據
filtered_data = merged_df2[merged_df2['統計期'].isin(['112年 8月'])]
filtered_data.head()

x = filtered_data['類別']
y = filtered_data['數量']

# 創建一個長條圖
plt.figure(figsize=(10, 6))  # 設置圖形大小
plt.bar(x, y, color='Blue')
plt.title('112年8月大重普重小客車的領牌情況')
plt.xlabel('類別')
plt.ylabel('數量')

# 顯示圖表
plt.tight_layout()
plt.show()
```
